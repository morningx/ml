## 分组聚合

- 按会员ID汇总
- 计算最近一次订单时间
- 计算订单频率

```python
rfm = data_merge.groupby(['year','会员ID'],as_index=False).agg({'data_interval':'min', # 最近一次订单时间
                                                             '订单号':'count',  # 频率
                                                             '订单金额':'sum'})  # 总金额
rfm.head()
# 每一年，每个会员，计算最小值间隔日期，订单总数，总金额
```

```python
# 重命名
# R值：值越小，等级越高
# F和M值：值越大，等级越高 
rfm.columns  = ['year','会员ID','r(最近)','f(购买数)','m(总金额)']
rfm.head()
```

# 特征权重-随机森林

- RandomForestClassifier(随机森林检测每个特征的重要性及每个样例属于哪个类的概率)

https://github.com/morningx/files/blob/master/MachineLearning/%E9%9A%8F%E6%9C%BA%E6%A3%AE%E6%9E%97/%E9%9A%8F%E6%9C%BA%E6%A3%AE%E6%9E%97%E5%8F%98%E9%87%8F%E9%87%8D%E8%A6%81%E6%80%A7-var%E9%87%8D%E8%A6%81%E6%80%A7-%E7%B2%BE%E7%A1%AE%E7%8E%87%E5%8A%A0%E5%99%AA%E9%9F%B3%E5%90%8E%E4%B8%8B%E9%99%8D%E9%87%8F-gini%E9%87%8D%E8%A6%81%E6%80%A7-%E4%B8%8D%E7%BA%AF%E5%BA%A6%E4%B8%8B%E9%99%8D%E7%A8%8B%E5%BA%A6%E4%B9%8B%E5%92%8C.png

```python
# 随机森林树计算字段权重 feature_importances
# random_state随机种子 可固定运行结果
# fit适配X参数和Y标签结果即可得到变量权重
clf = RandomForestClassifier(random_state=0)
clf.fit(rfm_merge[['r(最近)','f(购买数)','m(总金额)']],rfm_merge['会员等级'])
```

```python
# weights权重；重物；法码
# r最近消费天数 f频次 m金额
# 钱买的多权重高，其次是最近消费天数距离最后样本时间，频次特征不那么重要
weights = clf.feature_importances_
weights
# 间隔天数，频次，金额
# 金额重要性最大，其实是最近购买时间，最差为频次
```

# RFM正反比关系赋予权重

- array([0.41157118, 0.0059908 , 0.58243802])
- ['r(最近)','f(购买数)','m(总金额)']
- 金额多，则权重高，购买多则权重高，越近值越小间隔天数小权重大
- 对应会员 天数小、金额多、次数多，等级高，用户价值更大
- 天数反比，金额正比，次数正比

# 总得分=特征权重得分求和

```python
# category分类；种类；范畴 分类值变int值可用于计算
rfm = rfm.apply(np.int32)
rfm.info()
```

```python
weights[0],weights[1],weights[2]
# r,f,m
```

```python
rfm['rfm_score'] = rfm['r_socre']*weights[0]+rfm['f_score']*weights[1]+rfm['m_score']*weights[2]
# 随机森林对应的权重值，分箱区间不同对应价值不一样
# 综合样本的整体权重和分箱占比，让更有价值的字段对应的分箱更加明显
rfm.head()
```

```python
rfm.rename(columns={'r_socre':'r_score'},inplace=True) 
rfm.head(2)
```

# RFM模型分组聚合

```python
rfm_0 = rfm.copy()
rfm_0 = rfm_0.iloc[:,:-1]
# 获取所有行，除了Y标签得分的X特征
rfm_0.head()
```

```python
rfm_0['r_score'] = rfm_0['r_score'].astype(np.str)
rfm_0['f_score'] = rfm_0['f_score'].astype(np.str)
rfm_0['m_score'] = rfm_0['m_score'].astype(np.str)
```

```python
rfm_0['r_score'].str.cat(rfm_0['f_score']).str[0]
```

```python
# 用字符指针实现函数str_cat(s,t),将字符串t复制到字符串s的末端
rfm_0['rfm_group'] = rfm_0['r_score'].str.cat(rfm_0['f_score']).str.cat(rfm_0['m_score'])
rfm_0.head(2)
```
